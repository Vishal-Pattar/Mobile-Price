```python
def plot_relation_with_target(x,y):
    fig, ax = plt.subplots(1,2,figsize=(12,5))
    fig.suptitle(y + " v/s " + x)
    sns.barplot(data=train, x=x, y=y, ax=ax[0])
    sns.lineplot(data=train, x=x, y=y, ax=ax[1])
    plt.show()
```

# Battery Power V/S Price Range

```python
plot_relation_with_target('price_range','battery_power')
```

Observations:
- The price of mobile increases with increase in battery power
- The battery power is similar for the price range 1 and 2

# Bluetooth V/S Price Range

```python
plot_relation_with_target('blue','price_range')
```

Observations:
- Mobile price slightly increases with bluetooth module

# Clock Speed V/S Price Range

```python
plot_relation_with_target('price_range','clock_speed')
```

Observations:
- There is no clear pattern for the relation between price range and clock speed
