```python
fig, ax = plt.subplots(1,1,sharex=True, figsize=(15,12))
sns.heatmap(train.corr(), fmt='.2f', cmap="YlGnBu", annot=True)
```

# Battery Power V/S Price Range

```python
fig, ax = plt.subplots(1,2,figsize=(15,6))
sns.barplot(data=train, x='price_range', y='battery_power', ax=ax[0])
sns.lineplot(data=train, x='price_range', y='battery_power', ax=ax[1])
```

Observations:
- The price of mobile increases with increase in battery power
- The battery power is similar for the price range 1 and 2

# Bluetooth V/S Price Range
