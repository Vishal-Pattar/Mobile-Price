```python
fig, ax = plt.subplots(1,1,sharex=True, figsize=(15,12))
sns.heatmap(train.corr(), fmt='.2f', cmap="YlGnBu", annot=True)
```

Observations: \

2) px_height

4) ram

```python
posx = -1
j = -1
fig, ax = plt.subplots(10,4,figsize=(40,80))
for i in range ((len(columns)-1)*2):
    posy = i % 4
    if i % 4 == 0:
        posx += 1
    if i % 2 == 0:
        j += 1
    sns.barplot(data=train, x=columns[20], y=columns[j], ax=ax[posx,posy])
    sns.lineplot(data=train, x=columns[20], y=columns[j], ax=ax[posx,posy])
```
